# Video Unshuffle Environment Demo

This notebook demonstrates the Video Unshuffle environment where you reorder shuffled video frames.

```python
import gymnasium as gym
import numpy as np
from IPython.display import display
from PIL import Image
```

## 1. Create and Inspect Environment (Easy - 4 frames)

```python
env = gym.make("video_unshuffle/easy", seed=42)
print(f"Number of frames: {env.num_frames}")
print(f"Action Space: {env.action_space}")
print(f"Available actions: {env.action_space.get_function_names()}")
```

## 2. Reset and View Initial State

```python
obs, info = env.reset()
print(f"Observation shape: {obs.shape}")
print(f"Current order: {env.current_order if hasattr(env, 'current_order') else 'N/A'}")
print(f"Target order: {env.target_order if hasattr(env, 'target_order') else 'N/A'}")

display(Image.fromarray(obs))
print(f"\nTask: Reorder the shuffled frames to restore the original video sequence")
print(f"\nTask Prompt:\n{env.get_prompt()}")
```

## 3. Test Manual Actions

```python
# Test swapping or reordering frames (exact format depends on implementation)
print("Available actions depend on environment implementation (swap/reorder)")
```

## 4. Test Solver - Reorder Strategy

```python
env = gym.make("video_unshuffle/easy", seed=100)
obs, info = env.reset()
print("Initial shuffled frames:")
print(f"Current order: {env.current_order if hasattr(env, 'current_order') else 'N/A'}")
print(f"Target order: {env.target_order if hasattr(env, 'target_order') else 'N/A'}")
display(Image.fromarray(obs))

# Get solver actions (reorder strategy)
actions = env.solve(strategy="reorder") if hasattr(env, 'solve') else []
print(f"\nSolver actions: {actions}")

# Execute solver actions
for i, action in enumerate(actions):
    obs, reward, terminated, truncated, info = env.step(action)
    print(f"\nStep {i+1}: {action}")
    print(f"Current order: {env.current_order if hasattr(env, 'current_order') else 'N/A'}")
    display(Image.fromarray(obs))
    if terminated:
        print(f"\n✓ Video unshuffled! Reward: {reward}")
        break

env.close()
```

## 5. Test Solver - Swap Strategy

```python
env = gym.make("video_unshuffle/easy", seed=200)
obs, info = env.reset()
print("Initial state:")
display(Image.fromarray(obs))

# Get solver actions using swap strategy
if hasattr(env, 'solve'):
    actions = env.solve(strategy="swap", num_steps=5)
    print(f"\nSolver generated {len(actions)} swap actions")
    
    for action in actions:
        obs, reward, terminated, truncated, info = env.step(action)
        if terminated:
            print(f"Reward: {reward}")
            display(Image.fromarray(obs))
            break

env.close()
```

## 6. Test Hard Difficulty (5 frames)

```python
env_hard = gym.make("video_unshuffle/hard", seed=42)
obs, info = env_hard.reset()
print(f"Hard difficulty - {env_hard.num_frames} frames (vs easy: 4)")
print(f"Current order: {env_hard.current_order if hasattr(env_hard, 'current_order') else 'N/A'}")
display(Image.fromarray(obs))

# Get solution
if hasattr(env_hard, 'solve'):
    actions = env_hard.solve(strategy="reorder")
    print(f"\nSolution: {actions}")
    for action in actions:
        obs, reward, terminated, truncated, info = env_hard.step(action)
        if terminated:
            print(f"Reward: {reward}")
            break

env_hard.close()
```
